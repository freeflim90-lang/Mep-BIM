```python
# (1) Function Descriptions

# initialize_bems: This function initializes the energy management system with given grid and renewable prices.
# It also sets initial values for energy usage and total costs.
def initialize_bems(grid_price, renewable_price):
    return {
        "grid_price": grid_price,  # Cost per kWh for grid energy
        "renewable_price": renewable_price,  # Cost per kWh for renewable energy
        "grid_energy_used": 0,  # Total grid energy used in kWh
        "renewable_energy_used": 0,  # Total renewable energy used in kWh
        "total_energy_consumed": 0,  # Total energy consumed (grid + renewable)
        "total_cost": 0  # Total cost of energy consumed
    }

# use_renewable_energy: This function adds the renewable energy consumed to the total and calculates the cost.
def use_renewable_energy(system, energy):
    system["renewable_energy_used"] += energy  # Add to renewable energy total
    system["total_energy_consumed"] += energy  # Add to total energy consumed
    system["total_cost"] += energy * system["renewable_price"]  # Add cost for renewable energy used

# use_grid_energy: This function adds the grid energy consumed to the total and calculates the cost.
def use_grid_energy(system, energy):
    system["grid_energy_used"] += energy  # Add to grid energy total
    system["total_energy_consumed"] += energy  # Add to total energy consumed
    system["total_cost"] += energy * system["grid_price"]  # Add cost for grid energy used

# consume_energy: This function simulates energy consumption for an hour.
# If the building is occupied, it uses both renewable and grid energy. 
# If not, no energy is consumed.
def consume_energy(system, renewable_energy, grid_energy, occupied):
    if occupied:
        use_renewable_energy(system, renewable_energy)  # Use renewable energy
        use_grid_energy(system, grid_energy)  # Use grid energy
    else:
        pass  # No energy is consumed if the building is unoccupied

# calculate_total_cost: This function calculates and returns the total cost of energy used.
def calculate_total_cost(system):
    return system["total_cost"]

# generate_hourly_data: This function generates energy consumption data for 24 hours.
# Occupied is set to True for hours between 8:00 and 18:00 (10 hours of occupancy).
def generate_hourly_data():
    data = []
    for hour in range(24):  # Loop over 24 hours
        renewable_energy = 5  # Example value: 5 kWh of renewable energy per hour
        grid_energy = 3       # Example value: 3 kWh of grid energy per hour
        occupied = hour >= 8 and hour <= 18  # Building is occupied between 8:00 and 18:00
        data.append((renewable_energy, grid_energy, occupied))
    return data

# (2) Explanation of the Final Output
# The final output results from energy being consumed only during occupied hours (8 AM to 6 PM).
# During those hours, both renewable and grid energy are used. Outside of those hours, the building is unoccupied, 
# and no energy is consumed, leading to reduced energy consumption and costs.

# (3) Modify the simulation to run over a week (7 days)

# generate_weekly_data: This function extends the simulation to run for 7 days (a week).
def generate_weekly_data():
    data = []
    for day in range(7):  # Loop over 7 days
        for hour in range(24):  # Loop over 24 hours per day
            renewable_energy = 5  # Example value: 5 kWh of renewable energy per hour
            grid_energy = 3       # Example value: 3 kWh of grid energy per hour
            occupied = hour >= 8 and hour <= 18  # Building is occupied from 8:00 to 18:00
            data.append((renewable_energy, grid_energy, occupied))
    return data

# Simulate for one week
grid_price = 0.2  # $/kWh for grid energy
renewable_price = 0.1  # $/kWh for renewable energy

# Initialize BEMS system for a week
bems_week = initialize_bems(grid_price, renewable_price)

# Generate simulation data for a week
weekly_data = generate_weekly_data()

# Perform the simulation for the week
for renewable_energy, grid_energy, occupied in weekly_data:
    consume_energy(bems_week, renewable_energy, grid_energy, occupied)

# Calculate and display the results for the week
total_cost_week = calculate_total_cost(bems_week)
results_week = {
    "Total cost of energy used": f"${total_cost_week:.2f}",
    "Total renewable energy used": f"{bems_week['renewable_energy_used']} kWh",
    "Total grid energy used": f"{bems_week['grid_energy_used']} kWh",
    "Total energy consumed": f"{bems_week['total_energy_consumed']} kWh"
}

# Display the results
print(f"Total cost of energy used: {results_week['Total cost of energy used']}")
print(f"Total renewable energy used: {results_week['Total renewable energy used']}")
print(f"Total grid energy used: {results_week['Total grid energy used']}")
print(f"Total energy consumed: {results_week['Total energy consumed']}")
```
